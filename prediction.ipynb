```python
from tensorflow.keras import models
import cv2
from tensorflow.keras.applications.vgg16 import preprocess_input
import numpy as np
import matplotlib.pyplot as plt


class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# load the model
model = models.load_model('./VGG16_model.hdf5')

# load the image
o_img = cv2.imread('./bird.jfif')    
# resize the image to (32, 32)
img = cv2.resize(o_img, (32, 32))

X_ = np.array([img])
y_pred = np.argmax(model.predict(X_),axis = 1)

print(class_names[y_pred[0]])

plt.imshow(o_img)
plt.xlabel(class_names[y_pred[0]])
plt.show()
```
